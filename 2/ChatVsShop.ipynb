```python
import speech_recognition as sr
from gtts import gTTS
import pygame
import json
import os

# Kh·ªüi t·∫°o pygame mixer
pygame.mixer.init()

print("‚úÖ ƒê√£ import th√†nh c√¥ng c√°c th∆∞ vi·ªán")
```

```python
with open('qa_data.json', 'r', encoding='utf-8') as f:
    qa_data = json.load(f)

print(f"ƒê√£ load {len(qa_data)} c√¢u h·ªèi t·ª´ file JSON")
```

```python
def speak(txt):
    """Chuy·ªÉn text th√†nh gi·ªçng n√≥i - d√πng pygame"""
    try:
        # D·ª´ng √¢m thanh c≈© n·∫øu ƒëang ph√°t
        if pygame.mixer.music.get_busy():
            pygame.mixer.music.stop()
        pygame.mixer.music.unload()
        
        # T·∫°o file mp3 m·ªõi
        tts = gTTS(txt, lang='vi')
        tts.save('tmp.mp3')
        
        # Ph√°t √¢m thanh b·∫±ng pygame
        pygame.mixer.music.load('tmp.mp3')
        pygame.mixer.music.play()
        
        # Ch·ªù ph√°t xong
        while pygame.mixer.music.get_busy():
            pygame.time.Clock().tick(10)
            
    except Exception as e:
        print(f"L·ªói khi ph√°t √¢m thanh: {e}")

def get_answer(question):
    """T√¨m c√¢u tr·∫£ l·ªùi t·ª´ JSON"""
    question = question.lower().strip()
    
    # T√¨m ch√≠nh x√°c
    if question in qa_data:
        return qa_data[question]
    
    # T√¨m g·∫ßn ƒë√∫ng
    for key, value in qa_data.items():
        if key in question:
            return value
    
    return "Xin l·ªói, t√¥i ch∆∞a hi·ªÉu c√¢u h·ªèi n√†y. B·∫°n c√≥ th·ªÉ h·ªèi c√°ch kh√°c kh√¥ng?"

print("‚úÖ ƒê√£ ƒë·ªãnh nghƒ©a c√°c h√†m")
```

```python
print("B·∫Øt ƒë·∫ßu voice assistant...")
print("N√≥i 'ngh·ªâ' ho·∫∑c 't·∫°m bi·ªát' ƒë·ªÉ d·ª´ng\n")

while True:
    r = sr.Recognizer()
    
    try:
        with sr.Microphone() as source:
            print("\nüé§ Say something!")
            r.adjust_for_ambient_noise(source, duration=0.5)
            audio = r.listen(source, timeout=5, phrase_time_limit=10)
        
        # Nh·∫≠n di·ªán gi·ªçng n√≥i
        question = r.recognize_google(audio, language="vi-VN").lower()
        print(f"üë§ You said: {question}")
        
        # Ki·ªÉm tra l·ªánh d·ª´ng
        if 'ngh·ªâ' in question or 't·∫°m bi·ªát' in question:
            print("\n‚õî STOP")
            ans = "T·∫°m bi·ªát. H·∫πn g·∫∑p l·∫°i!"
            print(f"ü§ñ Answer: {ans}")
            speak(ans)
            break
        
        # T√¨m c√¢u tr·∫£ l·ªùi
        ans = get_answer(question)
        print(f"ü§ñ Answer: {ans}")
        speak(ans)
        
    except sr.WaitTimeoutError:
        print("‚è±Ô∏è Timeout - kh√¥ng nghe th·∫•y g√¨")
        continue
    except sr.UnknownValueError:
        print("‚ùå Kh√¥ng nghe r√µ")
        ans = 'T√¥i kh√¥ng nghe r√µ. B·∫°n n√≥i l·∫°i ƒë∆∞·ª£c kh√¥ng?'
        speak(ans)
    except sr.RequestError as e:
        print(f"‚ùå L·ªói: {e}")
        ans = "Xin l·ªói, c√≥ l·ªói x·∫£y ra!"
        speak(ans)
    except KeyboardInterrupt:
        print("\n\n‚õî ƒê√£ d·ª´ng b·ªüi ng∆∞·ªùi d√πng")
        break
    except Exception as e:
        print(f"‚ùå L·ªói kh√¥ng x√°c ƒë·ªãnh: {e}")
        continue

print("\n‚úÖ Voice Assistant ƒë√£ d·ª´ng")
```
